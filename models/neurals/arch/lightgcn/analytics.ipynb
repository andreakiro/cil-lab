```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('logs-50.pickle', 'rb') as handle:
    logs = pickle.load(handle)
```

```python
train = []
eval = []
for epochs in logs:
    train.append(logs[epochs]['train_loss'])
    eval.append(logs[epochs]['eval_loss'])
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(list(logs.keys()), train, label='train')
ax.plot(list(logs.keys()), eval, label='eval')
ax.legend()
```
